Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer

```

Data Pre-Processing

```python
#  Loading the dataset
mail_data = pd.read_csv('/content/mail data.csv')
mail_data
```

```python
# Information about the dataset
mail_data.info()
```

```python
# Dimension of the dataset
mail_data.shape
```

```python
mail_data.describe()
```

```python
# Replacing yhe null values with a null String
data = mail_data.where((pd.notnull(mail_data)),'')
```

```python
# printing the first 5 rows  of the df
data.head()
```

```python
# Label spam mail as 0 and ham  mail as 1
data.loc[data['Category'] == 'spam','category',] = 0
data.loc[data['Category'] == 'ham','category',] = 1
```

```python
# Separating the data as texts and label
X = data['Message']
Y = data['category']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Feature Extraction

```python
# Transform the test data to feature vectors that can used as input to the Logistic regreesion
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase =True)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as intergers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')


```

```python
print(X_train_features)
```

Training the Model

```python
# Training the Logistic Regression model with the training dataset
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

Model Evaluation

```python
#Accuracy score for training data
X_train_prediction = model.predict(X_train_features)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy score of training data: ",training_data_accuracy)
```

```python
#Accuracy score for testing data
X_test_prediction = model.predict(X_test_features)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy score of testing data",testing_data_accuracy)
```

Buliding a Predictive System

```python
#Making the Predictive system
input_mail = ["WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only."]

#Converting text into feature vectors
input_data_features = feature_extraction.transform(input_mail)

#Making prediction
prediction = model.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
  print('Ham mail')
else:
  print('Spam mail')
```
