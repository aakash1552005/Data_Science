```python

# @title converting csv into dataframes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('netflix.csv', lineterminator = '\n')
df
```

```python
df.describe()
```

```python
# @title Removing Null values
df.dropna(inplace= True)
df.isna ().sum()
```

```python
df.head()
```

```python
# @title 1) Converting date format into year wise?
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtypes)
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
df['Release_Date'].dtypes
```

```python
df.rename(columns={'Release_Date': 'Release_Year'}, inplace=True)
```

```python
df
```

```python
# @title 2) Dropping the unwanted columnns for our Analysis?
cols = ['Overview', 'Original_Language','Poster_Url']
```

```python
df.drop(cols, axis = 1, inplace = True)
df.columns
```

```python
df
```

```python
# @title 3) Categorising  Vote_Average into popular , average , below_avg and not_popular?

def catigorize_col(df, col, labels):
  edges = [df[col].describe()['min'],
           df[col].describe()['25%'],
           df[col].describe()['50%'],
           df[col].describe()['75%'],
           df[col].describe()['max']]

  df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
  return df
```

```python
labels = ['not_popular', 'below_avg', 'average', 'popular']
catigorize_col(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```

```python
df
```

```python
# @title 4) How many movies fall into each popularity category?
df['Vote_Average'].value_counts()
```

```python
# @title 5) What percentage of movies are "popular"?

(df['Vote_Average'].value_counts(normalize=True) * 100)
```

```python
# @title 6) Which year produced the most "not_popular" movies?
year = df[df['Vote_Average']=='not_popular']['Release_Year'].value_counts().idxmax()
count = df[df['Vote_Average']=='not_popular']['Release_Year'].value_counts().max()

print(year, count)
```

```python
# @title 7) Spilt the Genre into a list and then explode our dataframe to have only one Genre per row for each movie?

df['Genre'] = df['Genre'].str.split(', ')
df = df.explode('Genre').reset_index(drop=True)
df
```

```python
# @title 8)Finding number of unique values?
df.nunique()
```

```python
# @title 9) What movie got the highest popularity? What's its Genre?
df[df['Popularity']==df['Popularity'].max()]
```

```python
# @title 10) What movie got the Lowest popularity? What's its Genre?
df[df['Popularity']==df['Popularity'].min()]
```

```python
# @title 11) List all the Movies released before year 2020?
df[df['Release_Year']<2000]
```

```python
# @title 12) Which year released the highest number of movies?
df['Release_Year'].value_counts().idxmax()
```

```python
# @title 13) Which movie received the highest number of votes?
df.loc[df['Vote_Count'].idxmax(), ['Title', 'Vote_Count']]
```

```python
# @title 14) Which year added the most movies to Netflix?
df['Release_Year'].value_counts().sort_values(ascending=False)
```

```python
# @title 15) Popularity vs Vote_Average using  Scatter Plot?
plt.figure(figsize=(8,6))
sns.scatterplot(x='Vote_Average', y='Popularity', data=df)
plt.title("Popularity vs Rating")
plt.show()
```

```python
# @title 16) What is the most frequent Genre of movies released on Netflix?
sns.set_style('whitegrid')
sns.catplot(y = 'Genre', data = df, kind = 'count',
            order = df['Genre'].value_counts().index,
            color = '#4287f5')
plt.title('Genre column distribution')
plt.show()
```

```python
# @title 17) Which year has the most flimmed movies?
df['Release_Year'].hist()
plt.title('Release Date column distribution')
plt.show()
```

```python
# @title 18) Which  has the highest votes in Vote_Average column?
df['Vote_Average'].value_counts().plot(kind='bar')
```

```python
# @title 19) Heatmap of numeric correlations?
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# @title 20) Genre distribution pie chart?
genre_counts = df['Genre'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 10 Genres Distribution")
plt.show()
```
